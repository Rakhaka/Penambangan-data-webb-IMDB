```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link = 'https://www.imdb.com/chart/top/'
request = requests.get(link)
content = request.content
UserAgent = Request(link, headers={'User-Agent' : 'Mozilla/5.0'})
html_text = urlopen(UserAgent)
```

```python
soup = BeautifulSoup(html_text,"html.parser")
```

```python
daftar = soup.find_all("span", {"class": "sc-b189961a-1 kcfvgk"})
print(daftar)
```

```python
names = [tag.find_previous('a').text.strip() for tag in daftar]
rating = [tag.text.strip() for tag in daftar]
# tahun = [tag.find_previous('span', {'class': 'year'}).text.strip() for tag in daftar]
tahun = [tag.find_previous('div').text.strip() for tag in daftar]



df = pd.DataFrame({
    'Name': names,
    'Rating': rating,
    'Tahun' : tahun

})

print(df)
```

```python
# 1. menampilkan informasi 10 film teratas beserta rating IMDb-nya.
df.head(10)
```

```python
# 2. menentukan tahun rilis film dengan jumlah film terbanyak di daftar 100 film teratas.
tahun_rilis = df['Tahun'].value_counts()
print(tahun_rilis)
```

```python
# 3. menentukan jumlah film dengan rating IMDb di atas 9.0.
rating_terbanyak = df.groupby('Name')[['Rating']].sum()
rating_terbanyak = rating_terbanyak.sort_values('Rating', ascending = False) [:8]
rating_terbanyak
```
